## Bar charts and heatmaps

```python
# Flight delays by Airline
flight_data = pd.read_csv('data-for-datavis/flight_delays.csv', index_col="Month")
flight_data
```

```python
# Average arrival delay of NK airline (Spirit Airline)
plt.figure (figsize=(14,6))
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

sns.barplot (x=flight_data.index, y=flight_data['NK'])

plt.ylabel ('Arrival delay in minutes')
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

sns.heatmap (data=flight_data, annot=True)

plt.xlabel ('Airline')
```

## Scatter plots

```python
insurance_data = pd.read_csv('data-for-datavis/insurance.csv')
insurance_data.head(10)
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot (x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
# El gráfico sugiere una correlacion entre la prima del seguro y el bmi (indice de masa muscular).
# chequeamos la correlacion con una 'regression line'
plt.figure(figsize=(14,6))
sns.regplot (x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
# Podemos añadir la inforamación de un tercer campo y colorearlo diferente
plt.figure(figsize=(14,6))
sns.scatterplot(x=insurance_data.bmi, y=insurance_data.charges ,hue=insurance_data.smoker)
```

```python
# Podemos ver 2 regresiones con sns.lmplot
plt.figure(figsize=(14,6))
sns.lmplot (x='bmi', y='charges', hue='smoker', data=insurance_data)
```

```python
# Uso de los categorical scatter plots (usamos variables discretas en vez de continuas)
plt.figure(figsize=(14,6))
sns.swarmplot (x=insurance_data.smoker, y=insurance_data.charges)
```

## Distributions

```python
iris_data = pd.read_csv('data-for-datavis/iris.csv', index_col="Id")
iris_data.head()
```

```python
plt.figure(figsize=(14,6))
sns.distplot (a=iris_data['Petal Length (cm)'], kde=False)
```

The next type of plot is a <strong>kernel density estimate (KDE) plot</strong>.  
In case you're not familiar with KDE plots, you can think of it as a smoothed histogram.

```python
plt.figure(figsize=(14,6))
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

```python
# 2 dimensional KDE plot
plt.figure(figsize=(14,14))
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
```

```python
# Como cambiar el tamaño por defecto del jointplot
my_plot=sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
my_plot.fig.set_figwidth(10)
my_plot.fig.set_figheight(10)
```

```python
#Otra forma de cambiar el tamaño, pasando parametro height=x
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], 
              height=10, 
              kind="kde")
```

```python
# Color coded plots
iris_set_filepath = 'data-for-datavis/iris_setosa.csv'
iris_ver_filepath = 'data-for-datavis/iris_versicolor.csv'
iris_vir_filepath = 'data-for-datavis/iris_virginica.csv'

# Read the files into variables 
iris_set_data = pd.read_csv(iris_set_filepath, index_col="Id")
iris_ver_data = pd.read_csv(iris_ver_filepath, index_col="Id")
iris_vir_data = pd.read_csv(iris_vir_filepath, index_col="Id")

# Print the first 5 rows of the Iris versicolor data
iris_ver_data.head()
```

```python
plt.figure(figsize=(14,6))

sns.distplot (a=iris_set_data['Petal Length (cm)'], label='Iris setosa', kde=False)
sns.distplot (a=iris_ver_data['Petal Length (cm)'], label='Iris setosa', kde=False)
sns.distplot (a=iris_vir_data['Petal Length (cm)'], label='Iris setosa', kde=False)
plt.title ('Histogram of petal lenght by species')
plt.legend() # force legend to appear
```

```python
plt.figure(figsize=(14,6))

sns.kdeplot (data=iris_set_data['Petal Length (cm)'],label='Iris setosa',shade=True )
sns.kdeplot (data=iris_ver_data['Petal Length (cm)'],label='Iris setosa',shade=True )
sns.kdeplot (data=iris_vir_data['Petal Length (cm)'],label='Iris setosa',shade=True )
plt.title("Distribution of Petal Lengths, by Species")
```

## Choosing Plot Types and Custom Styles


```python
def plot():
    plt.figure(figsize=(14,3))

    sns.kdeplot (data=iris_set_data['Petal Length (cm)'],label='Iris setosa',shade=True )
    sns.kdeplot (data=iris_ver_data['Petal Length (cm)'],label='Iris setosa',shade=True )
    sns.kdeplot (data=iris_vir_data['Petal Length (cm)'],label='Iris setosa',shade=True )
    plt.title("Distribution of Petal Lengths, by Species")

sns.set_style("dark")
plot()
sns.set_style("darkgrid")
plot()
sns.set_style("whitegrid")
plot()
sns.set_style("white")
plot()
sns.set_style("ticks")
plot()
```
